```python
import pandas as pd, numpy as np
import os
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [25, 15]
plt.rcParams.update({'font.size': 20})

# Load data into dataframe
dirin = os.path.join("..", "..", "Data", "neutrons")
filein = "ExportAr41_SPSCylinder_FiberCoreEnabled_part1.csv"
# Load data into DataFrame
df = pd.read_csv(os.path.join(dirin, filein), index_col=False)
df = df[df.columns[:-1]]             # remove last empty col
df["r"] = (df.x**2 + df.y**2)**.5    # X-Y radius
```

# Example: track 1 high-energy gamma
In this notebook, we consider the event `744785` and track the gamma track `34`.

```python
selected_cols = ["tracknumber", "PID", "energydeposition", "kineticenergy", "time", "x", "y", "z", "r", "creatorprocess"]
event = df[df.eventnumber==744785]

print("[Info] These are all the entries of the gamma 34:")
event[event.tracknumber==34][selected_cols]
```

As you can see, the first entry has been written with `creatorprocess=nCapture` and has high-energy `~4 MeV`.

However, in the second entry (*that is very close in time*) the gamma has already lost `1.5 MeV`. Again, in the third entry, the gamma has lost further `1.5 MeV`.

```python
print("[Info] These are the secondary particles with gamma as Parent:")
event[event.ParentTrackID==34][selected_cols]
```

## Safety check: sum energy deposited in LAr < energy lost

```python
edep_by_cut_on_gamma = event[event.tracknumber==34].energydeposition.sum()
edep_by_secondaries = event[event.ParentTrackID==34].energydeposition.sum()

print("[Info] The sum of E deposited in LAr by the secondaries of this gammas is {:.2f} MeV".format(edep_by_cut_on_gamma + edep_by_secondaries))
```
