```python
import os, glob, numpy as np, pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [25, 15]
plt.rcParams.update({'font.size': 18})
import time
```

```python
results = pd.read_csv("rnd_search_tune_cnn.txt")
results = results[results.columns[1:]]
```

```python
sorted_results = results.sort_values("mean_test_score", ascending=False)
```

```python
sorted_results.columns
```

```python
import seaborn as sn

params_arch = ['param_n_1_conv_layers', 'param_n_1_max_pooling_layers', 
              'param_n_2_conv_layers', 'param_n_2_max_pooling_layers', 'param_n_3_dense_layers',]

params_layers = ['param_n_filters_1_conv_layers', 'param_n_filters_2_conv_layers', 
                 'param_kernel_size_1_conv_layers', 'param_kernel_size_2_conv_layers',   
                 'param_pool_size_2_max_pooling_layers', 'param_pool_size_1_max_pooling_layers',
                 'param_n_neurons_3_dense_layers']
              
params_activations = ['param_activation_1_conv_layers', 'param_activation_2_conv_layers', 
                      'param_activation_3_dense_layers', 'param_activation_f_layer']

params_optimizer = ['param_learningrate', 'param_l2', 'param_l1',              
                    'param_dropout', 'param_decay']

plt.title("Architecture")
for i, param in enumerate(params_arch):
    plt.subplot(len(params_arch), 1, i + 1)
    plt.scatter(results[param], results['mean_test_score'])
    plt.xlabel(param)
    plt.ylabel("Mean AUC Score - 3 Fold Cross Val")
plt.show()
```

```python
plt.title("Layers Details")
for i, param in enumerate(params_layers[2:]):
    plt.subplot(len(params_layers[2:]), 1, i + 1)
    plt.scatter(results[param], results['mean_test_score'])
    plt.title(param)
plt.show()
```
